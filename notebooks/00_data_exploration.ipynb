## nuScenes Data Structure

nuScenes organizes data in a relational graph:

```
Scene (log of 20s driving)
  └── Sample (keyframe, 2Hz → 40 samples/scene)
        ├── SampleData (sensor readings per sample)
        │     └── CAM_FRONT image: dashcam-style 1600×900 → resized to 672×448
        └── SampleAnnotation (3D boxes for all agents)
              ├── category: vehicle.car, human.pedestrian, etc.
              ├── visibility: token → 0, 1, 2, 3, 4 (0% → 100%)
              └── translation: [x, y, z] in ego vehicle frame
```

**Rarity scoring** (Phase 1a) assigns each sample a score 0–4 based on:
1. Proximity: any agent within 5m of ego vehicle path
2. Occlusion: any agent with 0–40% visibility
3. Density: ≥15 annotated 3D boxes in the scene
4. Weather: scene description contains rain/night/fog/storm

Only samples with score ≥ 3 are retained (~400–500 frames expected from trainval).
